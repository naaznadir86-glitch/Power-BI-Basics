#2.Name and explain three components of Power BI?
# Ans. The three main components of power BI work together to create,publish and consume reports and dashboards.
1. Power BI Desktop - runs on windows
2. Power BI Service - cloud based platform
3. Power BI Mobile App - view insights

In short

Desktop **BUILDS** > Service **SHARES** > Mobile **VIEWS**.

#3. Explain the Power BI workflow ?
# Ans. DATA > MODEL > VISUALISE > PUBLISH > SHARE
  1. **Connect to data **-- Import data from souces like excel,databases or cloud services.
  2.** Transform and Model** --clean,shape and relate data using power query and data modeling.
  3. **Create Reports** - Build visuals and reports in Power BI desktop.
  4. **Publish** - Upload reports to the Power BI service.
  5. **Share & View**  - Share dashboards and view them on web or mobile.

#6. Define data view,report view and model view .Explain the purpose of each view.
#Ans. **Data view** -- displays data in table format after it is loaded in Power BI
      **PURPOSE** -- View and verify data values,check column datatypes and         formats,create calculated columns.

      **Report View **--The main working area where reports are designed using visuals
      **PURPOSE** --Add filters & slicers.create charts,tables and dashboards,Arrange visuals for storytelling.

      **Model view** -- Shows relationships between tables in a diagram format
     **PURPOSE** -- Define data model structure.Improve data accuracy and performance.Create and manage table table relationships.
